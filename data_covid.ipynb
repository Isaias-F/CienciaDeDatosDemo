#### De 11 personas encuestadas 27.3% tiene 19 años de edad.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data_covid.csv', usecols=['¿Cuál es tu edad? (Sólo número)'])


values = data.value_counts().to_list()
name = data.value_counts().keys().to_list()
name = list(sum(name, ()))

print(data.value_counts().to_frame())
plt.pie(x=values, labels=name, autopct="%0.1f %%")
plt.title('Edades')
plt.show()
```
